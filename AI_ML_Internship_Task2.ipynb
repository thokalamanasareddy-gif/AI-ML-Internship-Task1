```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
```

```python
df = pd.read_csv('Titanic-Dataset.csv')

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(6,4))
plt.hist(df['Age'].dropna(), bins=20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Age'])
plt.title('Age Boxplot')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['number'])

plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(),
            annot=True,
            cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.pairplot(
    df[['Survived','Age','Fare','Pclass']]
)
plt.show()
```

```python
sns.countplot(
    x='Survived',
    data=df
)
plt.title('Survival Distribution')
plt.show()
```

```python
observations = """
1. Most passengers were between 20-40 years old.
2. Some outliers are present in Fare.
3. Fare and Pclass show moderate correlation.
4. Missing values exist in Age and Cabin columns.
5. Survival rate is lower than non-survival rate.
6. Data contains skewness in Fare distribution.
"""

with open("observations.txt", "w") as file:
    file.write(observations)

print("observations.txt created successfully!")
```
